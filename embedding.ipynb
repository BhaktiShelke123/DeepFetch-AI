## Embedding 
This ipynb file contains importing the OpenAI for text embedding (changing the dimensions) then the embedding technique is been stored in a  variable. New txt file is been imported using langchain document loaders, then the txt is been splitted into chunks using recursivecharactertextsplitter then 2 steps are been combined where in the chunks are been converted to vectors using embedding techniqque and stored in chromadb vector store database then a query is been passed and the document is been retrived using the similarity search.

#### Embedding
converting text into vectors


```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeddings
```

```python
text= "This is bhakti shelke" 
query_result= embeddings.embed_query(text)
query_result
```

```python
len(query_result)
# check the dimension of the vector
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-large",dimensions=1024)
embeddings_1024
```

```python
query="THis is bhakti shelke"
query_re = embeddings_1024.embed_query(query)
query_re
```

```python
len(query_re)
```

```python
from langchain_community.document_loaders import TextLoader
docs = TextLoader("speech.txt").load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_split = RecursiveCharacterTextSplitter(chunk_size=100, chunk_overlap=5)
text_split
```

```python
final_documents = text_split.split_documents(docs)
final_documents
```

# Vector Embedding and Vector StoreDB

```python
from langchain_community.vectorstores import Chroma
db=Chroma.from_documents(final_documents,embeddings_1024)
db
```

```python
text = "Healthcare and Innovation"
retriever_results = db.similarity_search(text)
print(retriever_results)
```
